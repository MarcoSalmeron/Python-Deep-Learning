# 2- Construcci√≥n de las RNA

```python
import tensorflow as tf
from tensorflow.keras import regularizers
from tensorflow.keras.models import Sequential
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras import Input
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.metrics import AUC, Precision, Recall
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, RobustScaler
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.base import BaseEstimator, TransformerMixin
```

## 2.1- Auxiliares

```python
def train_val_test_split(df, rstate=42, shuffle=True, stratify=None):
    strat = df[stratify] if stratify else None
    train_set, test_set = train_test_split(
        df, test_size=0.4, random_state=rstate, shuffle=shuffle, stratify=strat)
    strat = test_set[stratify] if stratify else None
    val_set, test_set = train_test_split(
        test_set, test_size=0.5, random_state=rstate, shuffle=shuffle, stratify=strat)
    return (train_set, val_set, test_set)
```

```python
def remove_labels(df, label_name):
    X = df.drop(label_name, axis=1)
    y = df[label_name].copy()
    return (X, y)
```

```python
num_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", RobustScaler())
])
```

```python
class CustomOneHotEncoder(BaseEstimator, TransformerMixin):

    def __init__(self):
        self._oh = OneHotEncoder()
        self._columns = None

    def fit(self, X, y=None):
        X_cat = X.select_dtypes(include=['object'])
        self._columns = pd.get_dummies(X_cat).columns
        self._oh.fit(X_cat)
        return self

    def transform(self, X, y=None):
        X_copy = X.copy()
        X_cat = X_copy.select_dtypes(include=['object'])
        X_num = X_copy.select_dtypes(exclude=['object'])
        X_cat_oh = self._oh.transform(X_cat)
        X_cat_oh = pd.DataFrame(X_cat_oh.toarray(),
                                columns=self._columns,
                                index=X_copy.index)
        X_copy.drop(list(X_cat), axis=1, inplace=True)
        return X_copy.join(X_cat_oh)
```

```python
class DataFramePreparer(BaseEstimator, TransformerMixin):

    def __init__(self):
        self._full_pipeline = None
        self._columns = None

    def fit(self, X, y=None):
        num_attribs = list(X.select_dtypes(exclude=['object']))
        cat_attribs = list(X.select_dtypes(include=['object']))
        self._full_pipeline = ColumnTransformer([
            ("num", num_pipeline, num_attribs),
            ("cat", CustomOneHotEncoder(), cat_attribs),
        ])
        self._full_pipeline.fit(X)
        self._columns = pd.get_dummies(X).columns
        return self

    def transform(self, X, y=None):
        X_copy = X.copy()
        X_prep = self._full_pipeline.transform(X_copy)
        return pd.DataFrame(X_prep,
                            columns=self._columns,
                            index=X_copy.index)
```

## 2.2- Divisi√≥n del conjunto de datos

```python
train_set, val_set, test_set = train_val_test_split(
    df, rstate=42, shuffle=True, stratify="diagnosed_diabetes"
)
```

```python
X_train, y_train = remove_labels(train_set, 'diagnosed_diabetes')
X_val, y_val = remove_labels(val_set, 'diagnosed_diabetes')
X_test, y_test = remove_labels(test_set, 'diagnosed_diabetes')
```

```python
print(f'Longitud del set de entrenamiento: {len(X_train)}')
print(f'Longitud del set de validacion: {len(X_val)}')
print(f'Longitud del set de test: {len(X_test)}')
```

## 2.3- RNA (perceptr√≥n multicapa)

```python
prep = DataFramePreparer()
```

```python
# preprocesar datos
X_train_prepared = prep.fit_transform(X_train)
X_val_prepared = prep.fit_transform(X_val)
X_test_prepared = prep.transform(X_test)
```

```python
# crear modelo
model = Sequential([
    Input(shape=(X_train_prepared.shape[1],)),
    Dense(64, activation="relu", kernel_regularizer=regularizers.l2(0.001)),
    Dropout(0.5),
    Dense(32, activation="relu", kernel_regularizer=regularizers.l2(0.001)),
    Dropout(0.3),
    Dense(1, activation="sigmoid")
])
```

```python
# compilar
model.compile(optimizer="adam",
              loss="binary_crossentropy",
              metrics=["accuracy", AUC(name="auc"), Precision(), Recall()])
```

```python
# entrenar modelo
es = EarlyStopping(monitor="val_loss", patience=3, restore_best_weights=True)

history = model.fit(X_train_prepared, y_train,
                    epochs=30,
                    batch_size=32,
                    validation_data=(X_test_prepared, y_test),
                    callbacks=[es])
```

## 2.4- M√©tricas del entrenamiento

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report
```

```python
history_df = pd.DataFrame(history.history)
```

```python
# accuracy
plt.figure(figsize=(8,5))
sns.lineplot(data=history_df[["accuracy", "val_accuracy"]])
plt.title("Accuracy vs Val Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend(["Train", "Validation"])
plt.show()

# loss
plt.figure(figsize=(8,5))
sns.lineplot(data=history_df[["loss", "val_loss"]])
plt.title("Loss vs Val Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend(["Train", "Validation"])
plt.show()
```

## 2.5- Evaluar Modelo

```python
val_loss, val_acc, val_auc, val_precision, val_recall = model.evaluate(X_val_prepared, y_val)

print("Val Loss:", val_loss)
print("Val Accuracy:", val_acc)
print("Val AUC:", val_auc)
print("Val Precision:", val_precision)
print("Val Recall:", val_recall)
```

```python
y_val_pred = model.predict(X_val_prepared)
```

```python
y_val_classes = (y_val_pred > 0.5).astype("int32")
```

```python
cm = confusion_matrix(y_val, y_val_classes)

plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=[0,1], yticklabels=[0,1])

plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")

plt.show()
```

```python
report = classification_report(y_val, y_val_classes, target_names=["Clase 0","Clase 1"], output_dict=True)

metrics = ["precision","recall","f1-score"]
classes = ["Clase 0","Clase 1"]

plt.figure(figsize=(8,5))
for i, metric in enumerate(metrics):
    scores = [report[c][metric] for c in classes]
    plt.bar(np.arange(len(classes)) + i*0.25, scores, width=0.25, label=metric)

plt.xticks(np.arange(len(classes)) + 0.25, classes)
plt.ylim(0,1.05)
plt.ylabel("Score")
plt.title("Classification Report Metrics")
plt.legend()

plt.show()
```
